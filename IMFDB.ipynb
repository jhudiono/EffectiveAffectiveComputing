## IMFDB annotations
    Expressions :   Anger, Happiness, Sadness, Surprise, Fear, Disgust
    Illumination  :   Bad, Medium, High
    Pose            :   Frontal, Left, Right, Up, Down
    Occlusion    :   Glasses, Beard, Ornaments, Hair, Hand, None, Others
    Age             :   Child, Young, Middle and Old
    Makeup       :   Partial makeup, Over-makeup
    Gender        :   Male, Female 
    
## [VGG-16](https://keras.io/applications/#vgg16) input requirements
* default input size = 224x224
* specify input shape if include_top = False. Has to be channels_last (224, 224, 3) or channels_first (3, 224, 224)
* Width and height no smaller than 48.

```python
# Scratch
path = "./data/IMFDB_final/SalmanKhan/Dabang2/*.txt"
file = glob(path)[0]
data = pd.read_csv(file, sep="\s+", header=None)
data['path'] = "./data/IMFDB_final/" + data[9] + "/" + data[7] + "/images" + data[2]
data[data['path'].isna()]
```

```python
def get_annotation_files():
    path = "data/IMFDB_final/*/*/*.txt"
    files = glob(path)
    pbar = tqdm_notebook(total=len(files))
    total_data = []
    for file in files:
        try:
            data = pd.read_csv(file, sep="\s+", header=None)#[[2, 7, 9, 10, 11, 13, 15]] 
            #data.columns = ['file', 'movie', 'name', 'gender', 'emotion', 'light', 'pose']
            data['path'] = "./data/IMFDB_final/" + data[9] + "/" + data[7] + "/images/" + data[2]
            total_data.append(data)
        except Exception as err:
            print("Skipping", file)
            print(err)
        pbar.update(1)
    return pd.concat(total_data)#[['emotion', 'gender', 'pose', 'path']]
```

```python

data = get_annotation_files()
print("# of files:", len(data))

def get_images(data):
    def get_image(path):
        try:
            p = Path(path)
            return p.is_file()
        except:
            return False # Image.open(path)
    data['image'] = data['path'].progress_apply(get_image)

get_images(data)
print(len(data))
print(len(data[data['image'] == True]))
print(len(data[data['image'] == False]))
```
